```python
from datetime import datetime
file.DATE=datetime.strptime('2014-7-1', '%Y-%m-%d')
plt.plot(file.DATE, file.TMIN,c = 'red', alpha=0.9)

plt.title('MINIMUM TEMPERARURE',fontsize=16)
plt.axis([0,58,1,10])
plt.xlabel('',fontsize=15)
plt.ylabel('temperature (c)',fontsize=16)
fig = plt.figure(dpi=120,figsize =(9,4))
fig.autofmt_xdate()
plt.legend(['temp 1','temp 2'] ,loc=2)
plt.text(10,50,'FIRST',fontsize=20, c = 'blue')
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```
